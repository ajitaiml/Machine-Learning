## Diabetes Predicitions

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
dataset
```

```python
dataset.keys()
```

```python
print(dataset['DESCR'])
```

```python
import pandas as pd
df_diabetes=pd.DataFrame(dataset.data,columns=['age',
  'sex',
  'bmi',
  'bp',
  's1',
  's2',
  's3',
  's4',
  's5',
  's6'])
df_diabetes.head()
```

```python
## Independent and Dependent Features
X = df_diabetes
y = dataset['target']
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
X_train.head()
```

```python
##Corrrelation
X_train.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

## Hyperparameter Tuning

```python
param = {
    'criterion': ['squared_error', 'friedman_mse', 'absoulte_error'],
    'splitter': ['best', 'random'],
    'max_depth': [1,2,3,4,5,10,15,20,25],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor = DecisionTreeRegressor()
```

```python
grid = GridSearchCV(estimator=regressor, param_grid=param, cv=5, n_jobs=-1,scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
## Prediction
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
print(r2_score(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```

```python
selected_model = DecisionTreeRegressor(criterion='squared_error', max_depth=3, max_features='sqrt', splitter='best')
```

```python
selected_model.fit(X_train, y_train)
```

```python
## Visualize the Decision Tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selected_model,filled=True)
```
