# Local Outlier Factor (LOCAL & GLOBAL)

```python
import pandas as pd
import numpy as np 
from sklearn.neighbors import LocalOutlierFactor
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_circles
X, _ = make_circles(n_samples=500, factor=0.5, noise=0.05, random_state=42)
```

```python
X = np.concatenate([X,np.random.uniform(low=-6,high=6,size=(10,2))])
```

```python
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)  # Adjust contamination as needed
y_pred = lof.fit_predict(X)
```

```python
print("Outliers:", (y_pred == -1).sum())
```

```python
# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='coolwarm', edgecolor='k')
plt.title("Local Outlier Factor (LOF) Outlier Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.neighbors import LocalOutlierFactor

# Generate circular data
X, _ = make_circles(n_samples=500, factor=0.5, noise=0.05, random_state=42)

# Apply LOF
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)  # ~5% outliers
y_pred = lof.fit_predict(X)
outliers = y_pred == -1

# Plot results
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c='b', label='Inliers', s=20)
plt.scatter(X[outliers, 0], X[outliers, 1], c='r', label='Outliers', s=50, edgecolors='k')
plt.title("Local Outlier Factor on make_circles data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()
```
